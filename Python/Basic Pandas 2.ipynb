```python
import pandas as pd
import numpy as np

df = pd.DataFrame([[1.4,np.nan],[7.1,-4.5],[np.nan,np.nan],[0.75,-1.3]],index=['a','b','c','d'], columns=['one','two'])

print("Original Data:\n",df)

print("\nSum is\n",df.sum())
print("\nSum is(with skipNa)\n",df.sum( skipna=False)) #It will consider NA

print("Sum over axis 1\n",df.sum(axis=1))
```

```python
data = pd.Series([1,np.nan,3.5,np.nan,7,None])
print('original\n',data)
print('\nDrop NA\n',data.dropna())
```

```python
data = pd.DataFrame([[1,np.nan,3.5,np.nan,7,None],[1,np.nan,3.5,np.nan,None,None],[np.nan,np.nan,np.nan,np.nan,np.nan,None]])
print('original\n',data)
print('\nDrop NA\n',data.dropna()) # drops entire row if there is NaN in any column
print('\nDrop NA only if entire row is NA\n',data.dropna(how='all')) # drops entire row if there is NaN in any column
print('\nDrop only columns which has all NA\n',data.dropna(axis=1,how='all')) # drops entire column if there are NA's completely
print('\n\nWe can set threshold value for dropping..\n')
print("After threshold:\n", data.dropna())
print("After threshold:\n", data.dropna(axis='rows',thresh=3))
```

```python
print('\nUse Fill NA to fill Nan with desired value..\n')
print(data.fillna(0))

print('\nOriginal:\n',data)
print('\nforward fill..\n')
print(data.fillna(method='ffill'))

print('\nBackard fill..\n')
print(data.fillna(method='bfill'))
```
